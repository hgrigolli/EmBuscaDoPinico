```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
import math


```

```python
dados = [["Meio-Oeste",82.92,4.25],["Nordeste",82.56,3.47],["Sul",75.93,7.36],["Oeste",73.58,6.73]]   

print(dados)
```

```python
for dado in dados:
    media = dado[1]
    desvio_padrao = dado[2]
    x = np.linspace(media - 3*desvio_padrao, media + 3*desvio_padrao, 100)
    plt.plot(x,norm.pdf(x, media, desvio_padrao),label=dado[0])
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```
